```python
# Import the modules
import pandas as pd
from sklearn.model_selection import train_test_split
import tensorflow as tf
```

```python
# Read the scaled CSV file exported from "Heart_model_1.LogisticRegression.ipynb" into a Pandas DataFrame
df = pd.read_csv('Resources/heart_data_scaled.csv')
df.head()
```

```python
# Separate the datafram into target and features
y = df['num']
X = df.drop(columns='num')
```

```python
# Confirm the target and features variables
print(y.head())
X.head()
```

```python
# Split the data
X_train, X_test, y_train, y_test = train_test_split(X, 
                                                    y, 
                                                    random_state=1, 
                                                    stratify=y)
X_train.shape
```

```python
# Compile, Train, and Evaluate the Model
```

```python
# Define the model 
nn = tf.keras.models.Sequential()

# First hidden layer
nn.add(tf.keras.layers.Dense(units=40, activation="relu", input_dim=18))

# Second hidden layer
nn.add(tf.keras.layers.Dense(units=20, activation="relu"))

# Output layer
nn.add(tf.keras.layers.Dense(units=1, activation="sigmoid"))

# Check the structure of the model
nn.summary()
```

```python
# Compile the model
nn.compile(loss="binary_crossentropy", optimizer="adam", metrics=["accuracy"])
```

```python
# Train the model
fit_model = nn.fit(X_train, y_train, epochs=100)
```

```python
# Evaluate the model using the test data
model_loss, model_accuracy = nn.evaluate(X_test,y_test,verbose=2)
print(f"Loss: {model_loss}, Accuracy: {model_accuracy}")
```
